But before anything else, we can see that there are many NaN values in the data set and there is nothing more annoying than a NAN values for a data scientist.

```python
world_power_data.fillna(0, inplace = True)
```

```python
world_power_data.head()
```

## Traversing the dataset

#### 1. By Index

```python
world_power_data.loc[45:50] #return data rows from index no. 45 to 50
```

#### 2. By Column

```python
world_power_data[["country", "year", "coal_prod_change_pct", "coal_prod_change_twh" ,"gas_prod_change_pct"]].head() #show only certain columns
```

### Removing data for the year 2020

```python
twentytwenty_idx = world_power_data[world_power_data['year'] == 2020].index #get index for all rows of the year 2020
```

```python
world_power_data.drop(index = twentytwenty_idx, inplace = True) #drop the rows for the year 2020
```

```python
world_power_data.reset_index(drop = True, inplace = True) #reset the index
```

### Filtering data for only a few countries

```python
#our focus countries
focus_countries = ['United States', 'Canada', 'Russia', 'United Kingdom', 'China']
```

```python
#filter for focus countries
power_data = world_power_data[world_power_data["country"].isin(focus_countries)]
```

```python
power_data
```

This data starts all the way from the year 1990. 
Lets filter it to only show values from the last 2 decades

### Filtering data based on year

```python
power_data = power_data[power_data["year"] >= 2000] #return data for only years greater than or equal to 2000
```

```python
power_data
```

```python
#get column names with the word consumption in them
consumption_cols = [col for col in power_data.keys()[power_data.keys().str.contains('consumption')]]
```

```python
consumption_cols.append('country') #append country column
consumption_cols.append('year') #append year column
```

```python
consumption_cols
```

```python
power_data = power_data[consumption_cols]
```

```python
power_data.reset_index(drop = True, inplace = True)
power_data
```

Let's group by the data to calculate the total consumption of energy for each country in our list

```python
total_power_consumed = power_data.groupby(['country']).sum()
```

```python
total_power_consumed
```

The united states seems to be topping the charts for energy consumptin in every category.

We can save this data as a CSV file

```python
total_power_consumed.to_csv('Total_Power_Consumed.csv')
```
